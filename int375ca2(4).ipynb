```python
import numpy as np # for handling missing data
import pandas as pd # for data manipulation and cleaning
import matplotlib.pyplot as plt # for plotting data records
import seaborn as sns # for advanced visualization
df = pd.read_csv(r"C:\Users\payas\Downloads\educational_dataset_india_5000.csv")
```

```python
# View first few rows
df.head()
# Clean column names
df.columns = df.columns.str.lower().str.strip().str.replace(r'[^a-zA-Z0-9_]', '_', regex=True)
```

```python
# Check available columns
print("ðŸ§¾ Column names after cleaning:")
print(df.columns.tolist())
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Correlation Between Class Size and Marks
sns.scatterplot(data=df, x='no__of_students_in_class', y='marks_percentage', hue='education_level')
plt.title("Marks vs Class Size")
plt.show()
```

```python
#State-Wise Average Marks
state_avg = df.groupby('state_city')['marks_percentage'].mean().sort_values(ascending=False).head(10)
state_avg.plot(kind='bar', color='salmon')
plt.title("Top 10 States by Average Marks")
plt.ylabel("Average Marks (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
#Age Distribution of Students
sns.histplot(data=df, x='age', bins=20, kde=True)
plt.title("Age Distribution of Students")
plt.show()
```

```python
#Which Stream Has the Highest Average Marks?
stream_avg = df.groupby('stream_field_of_study')['marks_percentage'].mean().dropna().sort_values(ascending=False)
stream_avg.plot(kind='bar', color='purple')
plt.title("Average Marks by Stream")
plt.ylabel("Marks (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
#Outliers in Marks Using Boxplot
sns.boxplot(data=df['marks_percentage'])
plt.title("Outliers in Student Marks")
plt.show()
```
